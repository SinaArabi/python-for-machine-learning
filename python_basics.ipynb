## Python Functions

functions.png

```python
# A simple Python function to check
# whether x is even or odd
def evenOdd(x):
    ans = "even"
    if (x % 2 == 0):
        print("even")
    else:
        print("odd")
        ans = "odd"
    return ans

# Driver code to call the function
y = 5
ans_x = evenOdd(y)
print(y, " is ", ans_x)
# evenOdd(3)
```

## Example code

*write a code to print all prime numbers less than 100*

```python
def print_primes():
    for num in range(2, 100):  # Start from 2
        is_prime = True
        for i in range(2, int(num ** 0.5) + 1):  # Check divisibility up to sqrt(num)
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num, end=" ")

print_primes()
```
